```python
import librosa
import librosa.display
import IPython.display as ipd
y, sr = librosa.load('audio chaitu')
librosa.display.waveshow(y)
```

Question2 :Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to
the spoken words and silence between the words

```python
import numpy as np
```

```python
# Calculate length of the signal
signal_length = len(y)
print(f"Signal Length: {signal_length} samples") 
# Calculate magnitude range of the signal
magnitude_range = np.max(y) - np.min(y)
print(f"Magnitude Range: {magnitude_range}")
```

##Question3 : Take a small segment of the signal and play it. 

```python
from IPython.display import Audio
# Define the start and end time of the segment you want to play (in seconds)
start_time = 1.0
end_time = 4.0
# Extract the segment
segment = y[int(start_time * sr):
            int(end_time * sr)] 
# Play the segment
Audio(segment, rate=sr)
```

```python
librosa.display.waveshow(segment)

```
